Flipping Image

Import Module

1. PIL (Python Imaging Library) untuk membuka, memanipulasi, dan menyimpan gambar dalam berbagai format.
2. Matplotlib , untuk membuat visualisasi seperti grafik, plot, histogram, dll.

```python
from PIL import Image
import matplotlib.pyplot as plt
```

Import and Load Image

```python
image_path = "apel.jpg"
image = Image.open(image_path)
image_pixel = image.load()
width, height = image.size
```

```python
# menentukan nilai x0 dan y0
x0 = 796
y0 = 799
```

```python
# membuat citra baru
image_new = Image.new("RGB", (width, height))
```

Flipping Vertikal

```python
# melakukan flipping citra vertikal
for y in range(height):
        for x in range(width):

            x_new = -x + (2 * x0)

            # pastikan gak lewat batas gambar
            if 0 <= x_new < width:
                # ambil pixel gambar asli dan gunakan untuk ke pixel baru
                pixel_new = image.getpixel((x,y))
                image_new.putpixel((x_new, y), pixel_new)

image_new.save('flippingver_gambar' + '.jpg')
```

```python
# plot gambar hasil flip dan gambar asli
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].axis('on')

ax[1].imshow(image_new)
ax[1].set_title('Flipping Vertikal Image')
ax[1].axis('on')

plt.show()
```

Flipping Horizontal

```python
# melakukan flipping citra horizontal
for y in range(height):
        for x in range(width):

            y_new = -y + (2 * y0)

            # pastikan gak lewat batas gambar
            if 0 <= y_new < height:
                # ambil pixel gambar asli dan gunakan untuk ke pixel baru
                pixel_new = image.getpixel((x,y))
                image_new.putpixel((x, y_new), pixel_new)

image_new.save('flippinghor_gambar' + '.jpg')
```

```python
# plot gambar hasil flip dan gambar asli
fig, ax = plt.subplots(nrows=2, figsize=(5, 10))
ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].axis('on')

ax[1].imshow(image_new)
ax[1].set_title('Flipping Horizontal Image')
ax[1].axis('on')

plt.show()
```
