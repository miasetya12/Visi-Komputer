```python
import cv2
import numpy as np
import tkinter as tk
from tkinter import filedialog, messagebox
from PIL import Image, ImageTk
import pandas as pd

def rotasi_90_derajat_jarum_jam(matriks):
    baris, kolom, _ = matriks.shape
    matriks_rotasi = np.zeros((kolom, baris, matriks.shape[2]), dtype=matriks.dtype)
    for i in range(baris):
        for j in range(kolom):
            matriks_rotasi[j, baris-1-i] = matriks[i, j]
    return matriks_rotasi

def rotasi_90_derajat_berlawanan_jarum_jam(matriks):
    baris, kolom, _ = matriks.shape
    matriks_rotasi = np.zeros((kolom, baris, matriks.shape[2]), dtype=matriks.dtype)
    for i in range(baris):
        for j in range(kolom):
            matriks_rotasi[kolom-1-j, i] = matriks[i, j]
    return matriks_rotasi

def pencerminan_horizontal(matriks):
    baris, kolom, _ = matriks.shape
    matriks_pencerminan = matriks.copy()
    for i in range(baris):
        for j in range(kolom):
            matriks_pencerminan[baris-1-i, j] = matriks[i, j]
    return matriks_pencerminan

def pencerminan_vertikal(matriks):
    baris, kolom, _ = matriks.shape
    matriks_pencerminan = matriks.copy()
    for i in range(baris):
        for j in range(kolom):
            matriks_pencerminan[i, kolom-1-j] = matriks[i, j]
    return matriks_pencerminan

# def simpan_nilai_intensitas(gambar):
#     with open("intensitas_pixel.txt", "w") as f:
#         baris, kolom, _ = gambar.shape
#         for i in range(baris):
#             for j in range(kolom):
#                 R, G, B = gambar[i, j]
#                 f.write(f"f({i}, {j}): R({R}), G({G}), B({B})\n")
        
def simpan_nilai_intensitas_to_excel(gambar):
    data = []
    baris, kolom, _ = gambar.shape
    for i in range(baris):
        for j in range(kolom):
            R, G, B = gambar[i, j]
            data.append([f"f({i}, {j})", R, G, B])
    
    df = pd.DataFrame(data, columns=["Piksel", "R", "G", "B"])

    max_sheet_size = 1048576  # Maksimum ukuran lembar kerja Excel
    num_rows = df.shape[0]
    num_sheets = (num_rows - 1) // max_sheet_size + 1

    with pd.ExcelWriter("nilai_piksel.xlsx") as writer:
        for i in range(num_sheets):
            start_idx = i * max_sheet_size
            end_idx = min((i + 1) * max_sheet_size, num_rows)
            df_subset = df.iloc[start_idx:end_idx]
            sheet_name = f"Sheet_{i+1}"
            df_subset.to_excel(writer, sheet_name=sheet_name, index=False)

    print(f"Data nilai piksel telah disimpan ke dalam file Excel dengan {num_sheets} lembar kerja.")

def load_image():
    file_path = filedialog.askopenfilename()
    if file_path:
        try:
            image = cv2.imread(file_path)
            if image is not None:
                return image
            else:
                messagebox.showerror("Error", "Failed to load image. Please select a valid image file.")
        except Exception as e:
            messagebox.showerror("Error", f"Error occurred while loading image: {str(e)}")
    return None

def process_image():
    image = load_image()
    if image is not None:
        cv2.imshow("Original Image", image)
        cv2.waitKey(0)

        # # Simpan nilai intensitas gambar
        # simpan_nilai_intensitas(image)

        # Simpan nilai intensitas gambar ke dalam file Excel
        simpan_nilai_intensitas_to_excel(image)

        # Rotasi 90 derajat searah jarum jam
        rotated_clockwise = rotasi_90_derajat_jarum_jam(image)
        cv2.imshow("Rotated 90 Degrees Clockwise", rotated_clockwise)
        cv2.waitKey(0)

        # Rotasi 90 derajat berlawanan arah jarum jam
        rotated_counterclockwise = rotasi_90_derajat_berlawanan_jarum_jam(image)
        cv2.imshow("Rotated 90 Degrees Counterclockwise", rotated_counterclockwise)
        cv2.waitKey(0)

        # Pencerminan horizontal
        mirrored_horizontal = pencerminan_horizontal(image)
        cv2.imshow("Mirrored Horizontal", mirrored_horizontal)
        cv2.waitKey(0)

        # Pencerminan vertikal
        mirrored_vertical = pencerminan_vertikal(image)
        cv2.imshow("Mirrored Vertical", mirrored_vertical)
        cv2.waitKey(0)

        cv2.destroyAllWindows()

# Create GUI window
window = tk.Tk()
window.title("Image Processor")

# Create buttons
load_button = tk.Button(window, text="Load Image", command=process_image)
load_button.pack(pady=10)

# Run the GUI loop
window.mainloop()
```
